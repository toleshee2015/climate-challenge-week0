```python
from scipy.stats import zscore
import numpy as np

cols = ["T2M", "T2M_MAX", "T2M_MIN", "PRECTOTCORR", "RH2M", "WS2M", "WS2M_MAX"]

data = df[cols].apply(pd.to_numeric, errors='coerce')

z_scores = np.abs(zscore(data, nan_policy='omit'))

outlier_mask = (z_scores > 3)

outlier_rows = outlier_mask.any(axis=1)

print("Number of outlier rows:", outlier_rows.sum())
```

```python
df["Outlier"] = outlier_rows
```

```python
# Drop rows with >30% missing
threshold = int(0.7 * df.shape[1])
df = df.dropna(thresh=threshold)

# Forward fill remaining values
df.ffill(inplace=True)
```

```python
import matplotlib.pyplot as plt

monthly_temp = df.groupby("Month")["T2M"].mean()

monthly_temp.plot(title="Monthly Average Temperature (T2M)")
plt.xlabel("Month")
plt.ylabel("Temperature")
plt.show()

print("Warmest month:", monthly_temp.idxmax())
print("Coolest month:", monthly_temp.idxmin())
```

```python
monthly_rain = df.groupby("Month")["PRECTOTCORR"].sum()

monthly_rain.plot(kind="bar", title="Monthly Rainfall")
plt.xlabel("Month")
plt.ylabel("Rainfall")
plt.show()
```
